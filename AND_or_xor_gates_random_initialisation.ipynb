```python
import numpy as np
import matplotlib.pyplot as plt

# AND dataset
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[0],[0],[1]])

# sigmoid
def sigmoid(x):
    return 1/(1+np.exp(-x))

# random initialization
np.random.seed(0)
weights = np.random.randn(2,1)
bias = np.random.randn(1)

lr = 0.1

# training
for i in range(10000):
    z = np.dot(X,weights)+bias
    pred = sigmoid(z)

    error = pred - y
    dW = np.dot(X.T,error)
    db = np.sum(error)

    weights -= lr*dW
    bias -= lr*db

print("Predictions:\n",pred)

# decision boundary plot
plt.scatter(X[:,0],X[:,1],c=y[:,0])
x = np.linspace(-0.5,1.5,10)
y_line = -(weights[0]*x + bias)/weights[1]
plt.plot(x,y_line)
plt.title("AND Gate Decision Boundary")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

def sigmoid(x):
    return 1/(1+np.exp(-x))

np.random.seed(2)

# random initialization
W1 = np.random.randn(2,2)
b1 = np.random.randn(1,2)

W2 = np.random.randn(2,1)
b2 = np.random.randn(1,1)

lr = 0.1

for i in range(10000):

    # forward
    z1 = np.dot(X,W1)+b1
    a1 = sigmoid(z1)

    z2 = np.dot(a1,W2)+b2
    a2 = sigmoid(z2)

    # backprop
    error = a2-y
    dW2 = np.dot(a1.T,error)
    db2 = np.sum(error)

    hidden_error = np.dot(error,W2.T)*a1*(1-a1)
    dW1 = np.dot(X.T,hidden_error)
    db1 = np.sum(hidden_error)

    # update
    W2 -= lr*dW2
    b2 -= lr*db2
    W1 -= lr*dW1
    b1 -= lr*db1

print("Predictions:\n",a2)
```
