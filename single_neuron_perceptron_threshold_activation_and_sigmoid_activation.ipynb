```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler

# Load dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Use only 2 features for visualization
X = X[:, :2]

# Normalize
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Initialize weights
w = np.random.randn(2)
b = np.random.randn()
lr = 0.01
epochs = 100

# Threshold activation
def step(x):
    return 1 if x >= 0 else 0

# Training
for _ in range(epochs):
    for i in range(len(X)):
        z = np.dot(X[i], w) + b
        y_pred = step(z)

        error = y[i] - y_pred

        w += lr * error * X[i]
        b += lr * error

# Plot decision boundary
plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr')

x_vals = np.linspace(X[:,0].min(), X[:,0].max(), 100)
y_vals = -(w[0]*x_vals + b)/w[1]

plt.plot(x_vals, y_vals)
plt.title("Perceptron with Threshold Activation")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler

# Load dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Use 2 features for graph
X = X[:, :2]

# Normalize
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Initialize weights
w = np.random.randn(2)
b = np.random.randn()
lr = 0.01
epochs = 100

# Sigmoid activation
def sigmoid(z):
    return 1/(1+np.exp(-z))

# Training
for _ in range(epochs):
    for i in range(len(X)):
        z = np.dot(X[i], w) + b
        y_pred = sigmoid(z)

        error = y[i] - y_pred

        w += lr * error * X[i]
        b += lr * error

# Plot decision boundary
plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr')

x_vals = np.linspace(X[:,0].min(), X[:,0].max(), 100)
y_vals = -(w[0]*x_vals + b)/w[1]

plt.plot(x_vals, y_vals)
plt.title("Single Neuron with Sigmoid Activation")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
