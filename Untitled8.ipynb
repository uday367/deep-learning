```python
import numpy as np

# Input sequence
X = [1, 2, 3]

# Fixed weights
Wx = 0.4
Wh = 0.6
b = 0.2

# Initial hidden state
h = 0

print("RNN Forward Pass")

for t in range(len(X)):
    x = X[t]

    # Forward computation
    h = np.tanh(Wx*x + Wh*h + b)

    print("Step", t+1, "Hidden state =", h)
```

```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

# Input sequence
X = [1,2,3]

# Fixed weights
Wf, Uf, bf = 0.5, 0.4, 0.1
Wi, Ui, bi = 0.6, 0.5, 0.1
Wc, Uc, bc = 0.7, 0.3, 0.1
Wo, Uo, bo = 0.8, 0.2, 0.1

# Initial states
h = 0
C = 0

print("LSTM Forward Pass")

for t in range(len(X)):
    x = X[t]

    f = sigmoid(Wf*x + Uf*h + bf)
    i = sigmoid(Wi*x + Ui*h + bi)
    C_bar = np.tanh(Wc*x + Uc*h + bc)

    C = f*C + i*C_bar

    o = sigmoid(Wo*x + Uo*h + bo)
    h = o * np.tanh(C)

    print("Step",t+1)
    print("Hidden state =",h)
    print("Cell state =",C)
```
